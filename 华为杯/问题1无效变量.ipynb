```python
import pandas as pd
data_sum= pd.read_csv("train.csv",engine='python')
data_sum
```

```python
#得到缺失无效变量
columns=list(data_sum.columns)
lis=[]
for i in range(1,data_sum.shape[1]):
    col=data_sum.iloc[:,i]
    col1=col.apply(lambda x: abs(x))
    if col1.sum()==0:
        lis.append(columns[i])
lis
        
    
```

```python
ls1=pd.DataFrame(lis,columns=["缺失无效变量"])
#ls1.to_csv("缺失无效变量1.csv")
```

```python

f=open("缺失无效变量1.txt","w")
for n in lis:
    f.write(str(n))
    f.write(" ")
f.close()
    
```

```python
data_sum1=pd.DataFrame(data_sum)
data_sum1
data_sum1.drop(columns=lis,axis=1,inplace=True)  #axis=1表示删除列
data_sum1
```

```python
data_sum1
data_sum1.to_csv("删除后train.csv")
```
