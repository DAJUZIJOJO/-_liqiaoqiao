```python
def sed(n):
    import matplotlib as mpl
    import numpy as np
    import warnings 
    import matplotlib.pyplot as plt
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=n) #进行特征相关性分析
    #传统决策树线性回归,随机森林回归，极端森林回归
    rf1=MLPRegressor(max_depth=40,random_state=n)
    rf1.fit(X_train, y_train)
    rf2=RandomForestRegressor(n_estimators=100,max_depth=40,random_state=n)  #一般来说n_estimators越大越好，运行结果呈现出的两种结果该值分别是10和1000
    rf2.fit(X_train, y_train)
    rf3=ExtraTreesRegressor(n_estimators=100,max_depth=40,random_state=n)
    rf3.fit(X_train, y_train)
    predict1 = rf1.predict(X_test)
    predict2 = rf2.predict(X_test)
    predict3 = rf3.predict(X_test)
    score1= rf1.score(X_test, y_test) 
    result1= rf1.predict(X_test) 
    score2= rf2.score(X_test, y_test) 
    result2= rf2.predict(X_test) 
    score3= rf3.score(X_test, y_test) 
    result3= rf3.predict(X_test) 
    score1=str(score1)+","+str(score2)+","+str(score3)
    score=score1.split(",")
   # score.columns=["决策树回归精确度","随机森林回归精确度","极端森林回归精确度"]
    return score
```
